# 1. Create a function that takes a number as an argument and returns True or False depending on whether the number is symmetrical or not. A number is symmetrical when it is the same as its reverse.

```python
def is_symmetrical(number):
    list_number = [int(x) for x in str(number)]
    if list_number[::1] == list_number[::-1]:
        return True
    else:
        return False
```

```python
number = int(input("please enter number \t"))
```

```python
is_symmetrical(number)
```

# 2. Given a string of numbers separated by a comma and space, return the product of the numbers

```python
def multiply_nums(numbers):
    b = 1
    for i in numbers:
        b = b*i
    return b
```

```python
import re
mynum = str(input("please enter number with comma or space \t"))
mynum = re.split(',| ',mynum)
multiply_nums([int(x) for x in mynum])
```

# 3. Create a function that squares every digit of a number.

```python
def square_number(number):
    s = [str(i**2) for i in number]
    print(s)
    return int("".join(s))
```

```python
number = int(input("Please enter any number \t"))
square_number([int(x) for x in str(number)])
```

# 4.Create a function that sorts a list and removes all duplicate items from it.

```python
def list_sort(mynumber):
    myset = set(mynumber)
    mynumber = list(myset)
    mynumber.sort()
    return mynumber
```

```python
list_sort([3,4,2,6,8,11,113,99,11,123,44,55,44,678,908,678])
```

# 5.Create a function that returns the mean of all digits.

```python
def mean_digits(number):
    mean = 0
    count_number = len(number)
    for i in number :
        mean += i
    return mean // count_number
```

```python
number = int(input("Please enter numbers to calculate mean\t"))
mean_digits([int(x) for x in str(number)])
```
