```python
import csv
import mysql.connector as connect
```

```python
def carbon_logging(message):
    import logging as log
    log.basicConfig(filename='carbon_log.log',level=log.INFO,format="%(message)s %(asctime)s",datefmt="%Y-%m-%d %H:%M:%S")
    log.info(message)
```

```python
try:
    mycarbondb = connect.connect(host='localhost',username='root',password='mysql@007',use_pure=True)
except Exception as e:
    carbon_logging(e)
else:
    carbon_logging("DB is connected successfully")
```

```python
dbcursor = mycarbondb.cursor()
#### Creating database
query = "create database carbonnao"
try :
    dbcursor.execute(query)
except Exception as e:
    carbon_logging(e)
else:
    carbon_logging("carbonnano database is created successfully")
```

```python
#### Creating database table
query = """create table carbonnao.carbon_csv2(chiral_indice_N varchar(20), chiral_indice_M varchar(20), Initial_atomic_coordinate_U varchar(20),Initial_atomic_coordinate_V varchar(20),Initial_atomic_coordinate_W varchar(20), Calculated_atomic_coordinates_U varchar(20),
        Calculated_atomic_coordinates_V varchar(20),Calculated_atomic_coordinates_W varchar(20))"""
try :
    dbcursor.execute(query)
except Exception as e:
    carbon_logging(e)
else:
    carbon_logging("carbonnao.carbon_csv2 database table is created successfully")
```

```python
try:
    f = open("carbon_nanotubes.csv","r")
    f.seek(0)
    next(f)
except Exception as e:
    carbon_logging(e)
else:
    carbon_logging("File is object is created")   
```

```python
try:
    file_csv = csv.reader(f,delimiter=";")
except Exception as e:
    carbon_logging(e)
else:
    carbon_logging("CSV is created using delimiter")
    
for i in file_csv:
    i = str(i).strip('[')
    i = str(i).strip(']')
    query = "insert into carbonnao.carbon_csv2 values({})".format(str(i))
    try:
        dbcursor.execute(query)
    except Exception as e:
        carbon_logging(e)
    else:
        carbon_logging("{} is inserted successfully".format(i))
        mycarbondb.commit()
else:    
    carbon_logging("CSV file is successfully uploaded into database table")
```

```python
try:
    mycarbondb.close()
except Exception as e:
    carbon_logging(e)
else:
    carbon_logging("DB connection is closed")
```
