## 2. Write a Python Program to Find Factorial of Number Using Recursion?

```python
def factorial(num):
    if num == 1 :
        return 1
    else:
        return num*factorial(num-1)
```

```python
mynum = int(input("Please enter any number \t"))
print("The factorial of given number {} is {}".format(mynum,factorial(mynum)))
```

## 1. Write a Python Program to Display Fibonacci Sequence Using Recursion?

```python
def fibnocci_recursive(num):
    if num <= 1:
        return num
    else:
        return (fibnocci_recursive(num-1) + fibnocci_recursive(num-2))
```

```python
mynum = int(input("Please enter any number \t"))

for i in range(mynum):
    print(fibnocci_recursive(i))
```

## 3. Write a Python Program to calculate your Body Mass Index?

```python
def bmi(weight,height):
    return round(weight/(height**2),3)
```

```python
weight = float(input("Please enter your weight in kilograms(KGs) \t"))
height = float(input("Please enter your height in feet \t"))
height_meter = height / 3.281
print(f"BMI of your given weight {weight} and height {height} is {bmi(weight,height_meter)}")
```

## 5. Write a Python Program for cube sum of first n natural numbers?

```python
def find_cube(n):
    return n**3

def cube_natural(n):
    return map(find_cube,range(1,n))
```

```python
naturla_n = int(input('Please enter any number for find the sum of cube \t'))
print("The cube sum of natural number % d is %d"%(naturla_n,sum(cube_natural(naturla_n))))
```

## 4. Write a Python Program to calculate the natural logarithm of any number?

```python
def find_log(n):
    import math
    return math.log(n)
```

```python
n = int(input("Please enter number to find log \t"))
print("Logarith for given number %d is %f"%(n,find_log(n)))
```
