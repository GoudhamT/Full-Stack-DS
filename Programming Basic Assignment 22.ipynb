# 1.Create a function that takes three parameters where:
##  x is the start of the range (inclusive).
##  y is the end of the range (inclusive).
##  n is the divisor to be checked against.

```python
def get_range_divisible(x,y,z):
    output_lst = []
    for i in range(x,y+1):
        if i % z == 0:
            output_lst.append(i)
    return output_lst
```

```python
x = int(input("Please enter starting number \t"))
y = int(input("Please enter ending number \t"))
z = int(input("Please enter number to be divided \t"))
print("For given input range of {} is divisible by {}".format(get_range_divisible(x,y,z),z))
```

# 2. Create a function that takes in two lists and returns True if the second list follows the first list by one element, and False otherwise. In other words, determine if the second list is the first list shifted to the right by 1.

```python
num1 = str(input("Please enter more than one number \t"))
num2 = str(input("Please enter more than one number \t"))
if num1.isnumeric() and num2.isnumeric():    
    if len(num1) != len(num2):
        print("Your input should be same length")
    elif len(num1) < 2 or len(num2) < 2:
        print("Enter more than one number")
    else:
        print(compare_list(num1,num2))
else:
    print("Only number is allowed")
```

```python
def compare_list(num1,num2):
    list1  = [int(i) for i in num1]
    list2 = [int(j) for j in num2]
    list1.pop()
    list2.pop(0)
    if list1 == list2:
        return True
    else:
        return False
```

# 4. An isogram is a word that has no duplicate letters. Create a function that takes a string and returns either True or False depending on whether or not it's an "isogram

```python
def is_isogram(mystring):
    if len(mystring.lower()) == len(set(mystring.lower())):
           return True
    else:
           return False
```

```python
mystring = str(input("Enter any word to find isogram or not \t"))
is_isogram(mystring)
```

# 5. Create a function that takes a string and returns True or False, depending on whether the characters are in order or not.
## example: is_in_order(&quot;abc&quot;) âžž True

```python
def is_order(mystring):
    mystring.replace(' ','')
    highest = 0
    current = 0
    for i in mystring:
        current = ord(str(i))
        if current > highest:
            highest = current
        else:
            return False
    else:
        return True
        
```

```python
mystring = str(input("Please enter any word to find is order or not \t"))
is_order(mystring)
```

# 3. A group of friends have decided to start a secret society. The name will be the first letter of each of their names, sorted in alphabetical order.
## Create a function that takes in a list of names and returns the name of the secret society.

```python
def secret_society(names):
    names.sort()
    testl = [i[0] for i in names]
    print('created secret society is',''.join(testl))  
```

```python
names_list = []
input_ok = ''
while input_ok != 'yes':
    names = str(input("Please enter any name to form secret society \t"))
    names_list.append(names)
    input_ok = input("Please enter yes if you want to create society with given names, want to give more names then enter no\t")
else:
    secret_society(names_list)
```
