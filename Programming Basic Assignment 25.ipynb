# 1.Create a function that takes three integer arguments (a, b, c) and returns the amount of integers which are of equal value.

```python
#### This is traditional method
def count_letter(a,b,c):
    if (a == b) and (a == c):
        return 3
    elif (a == b) or (a == c):
        return 2
    elif (b == c ):
        return 2
    else:
        return 0
```

```python
a = int(input("Please enter any number \t"))
b = int(input("Please enter any number \t"))
c = int(input("Please enter any number \t"))

count_letter(a,b,c)
```

```python
#### Tried using list option
def count_letter_in_list(a,b,c):
    l = []
    l.append(a)
    l.append(b)
    l.append(c)
    equal = 0
    for i in l:
        equal = l.count(i)
        if equal == 3:
            break
        elif equal > 1 and equal < 3:
            break
        else:
            equal = 0
    return equal
        
```

```python
a = int(input("Please enter any number \t"))
b = int(input("Please enter any number \t"))
c = int(input("Please enter any number \t"))

count_letter_in_list(a,b,c)
```

# 2. Write a function that converts a dictionary into a list of keys-values tuples.

```python
def dic_to_tuples(mydict):
    l = list(mydict.items())
    l.sort()
    return l
```

```python
dic_to_tuples({"D":1,"B":2,"C":3})
```

```python
dic_to_tuples({"likes":2,"dislikes":3,"followers":10})
```

# 3. Write a function that creates a dictionary with each (key, value) pair being the (lower case,upper case) versions of a letter, respectively.

```python
def dict_upper_lower(mylist):
    mydict = {}
    for i in mylist:
        mydict[i.lower()] = i.upper()
    else:
        return mydict
        
```

```python
user_input = str(input("please enter only characters \t"))
if len(user_input) > 0:
    mylist = user_input.split(',')
    print(dict_upper_lower(mylist))
else:
    print("Please enter correct input")
```

# 4 . Write a function, that replaces all vowels in a string with a specified vowel.

```python
def vowel_replace(myinputstr,letter_replace):
    vowel_list = ['a','e','i','o','u']
    mydummystr = ""
    for i in myinputstr:
        if i.lower() in vowel_list:
            mydummystr += letter_replace
        else:
            mydummystr += i
    return mydummystr
```

```python
myinputstr = str(input("Please enter any words \t"))
letter_replace = str(input("Please enter any character to replace the vowel\t"))

vowel_replace(myinputstr,letter_replace)
# vowel_replace("apples and bananas","u")
```

# 5.Create a function that takes a string as input and capitalizes a letter if its ASCII code is even and returns its lower case version if its ASCII code is odd.

```python
def ascii_cap(myinputstr):
    myoutputstr = ""
    for i in myinputstr:
        if ord(i) % 2 == 0:
            myoutputstr += i.upper()
        else:
            myoutputstr += i.lower()
    return myoutputstr
            
```

```python
myinputstr = str(input("Please enter any words \t"))
ascii_cap(myinputstr)
```
