# 1 . Create a function that takes an integer and returns a list from 1 to the given number, where:
## 1. If the number can be divided evenly by 4, amplify it by 10 (i.e. return 10 times the number).
## 2. If the number cannot be divided evenly by 4, simply return the number.

```python
def mutiples_of4_list(number):
    l = []
    for i in number:
        if i % 4 == 0:
            l.append(i*10)
        else:
            l.append(i)
    return l
```

```python
class ValueTooSmallError(Exception):
    pass

number = int(input("enter any number"))
try : 
    if number >= 1:
        number = range(1,number+1)
        print(mutiples_of4_list(number))
    else:
        print("test")
        raise ValueTooSmallError
except ValueTooSmallError:
    print("This value is too small, try again!")
```

```python
#### Same output in list comprehension
number = int(input("enter any number"))
[ i * 10 if i % 4 == 0 else i for i in range(1,number+1)]
```

# 2. Create a function that takes a list of numbers and return the number that&#39;s unique.

```python
def get_unique(list_number):
    myset = set(list_number)
    print(myset)
    for i in myset:
        if list_number.count(i) == 1:
            return i
        
```

```python
number = str(input("Please enter number with comma separated \t"))
list_number = number.split(',')
final_result = get_unique(list_number)
print(final_result)
if final_result == None:
    print("Entered input is wrong")
else:
    print(f"for the give {number} set unique value is {final_result}")
```

# 4. Create a function that takes a list of strings and return a list, sorted from shortest to longest.

```python
def sort_lenght(mylist):
    return sorted(mylist,key=len)
```

```python
def is_Pythagorean(mylist):
    mylist = list(mylist)
    mylist.sort()
    min_sq = 0
    max_sq = mylist[-1]**2
    for i in mylist[0:2]:
        min_sq += i ** 2
    if min_sq == max_sq :
        return True
    else:
        return False
```

```python
is_Pythagorean((5,3,4))
```

```python
class circle():
    import math
    def __init__(self,radius):
        self.radius = radius
        
    def get_area(self):
        return round(math.pi *self.radius**2)
    
    def get_perimeter(self):
        return round(2*math.pi*self.radius)
        
```

```python
myradius = float(input("Please enter any value for radius of circle"))
```

```python
mycircle = circle(myradius)
```

```python
mycircle.get_area()
```
