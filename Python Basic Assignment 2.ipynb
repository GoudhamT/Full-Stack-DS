# 1. Write a Python program to convert kilometers to miles?
       
       ## Formula is miles = kilometer * .62317

```python
def get_miles(km):
    return km * .62317
```

```python
km = float(input("Please enter kilometer \n"))
```

```python
miles = get_miles(20)
print("For given kilometer",km,"miles is ",miles)
```

# 2. Write a Python program to convert Celsius to Fahrenheit?
  
     Fahrenheit = (Celsius * 9/5)+ 32

```python
def convert_cel_to_fh(cel):
    return (cel*(9/5))+32
```

```python
cel = float(input("Please enter temperature in celsius \n"))
```

```python
fh = convert_cel_to_fh(cel)
print(f"For given celsius temperature",cel,"Fahrenheit is {:.2f}".format(fh))
```

# 5. Write a Python program to swap two variables without temp variable?

```python
a = input("Enter a value \t")
b = input("Enter b value \t")
print("Variable a is %s"%a)
print("Variable b is %s"%b)
a,b = b,a
print("After swapping value of a is %s and b is %s"%(a,b))
```

# 4. Write a Python program to solve quadratic equation?

     Formula is (-b Â± (b ** 2 - 4 * a * c) ** 0.5) / 2 * a
     where a,b,c are real numbers and a cannot be 0

```python
def quadratic_eqn(a,b,c):
    import cmath 
    d = (b**2) - (4*a*c)
    sol1 = (-b-cmath.sqrt(d))/(2*a)
    sol2 = (-b+cmath.sqrt(d))/(2*a)
    return "The solution are", sol1, sol2
```

```python
x = int(input("Enter value a \t"))
if x == 0:
    print("You cannot enter value 0")
else:
    y = int(input("Enter value b \t"))
    z = int(input("Enter value c \t"))
    print(quadratic_eqn(x,y,z))
    

```
