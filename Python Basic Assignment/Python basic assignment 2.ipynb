# 10. How can you tell the difference between break and continue?

```python
mytext = 'PYTHON'

### Break - when condition is met, this will exit from loop
### continue - when condition is met, loop will continue again

### example scenario: when loop reaches N, exit from loop do not print anything, when loop meets T, do not print only T but continue to print remaining values
for i in mytext:
    if i == 'N':
        break
    elif i == 'T':
        continue
    else:
        print(i)
        
```
