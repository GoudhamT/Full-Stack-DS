# 1. Write a Python Program to Find LCM?

```python
"""This is old methodology to find LCM"""
def find_LCM(Number1, Number2):
    if Number1 > Number2:
        greater = Number1
    else:
        greater = Number2

    while True:
        if (greater % Number1 == 0) and (greater % Number2 == 0):
            lcm = greater
            break
        else:
            greater += 1
    return "LCM using general while condition is {}".format(lcm)
```

```python
"""This is using MATH package
formula is (number1 * number2) / GCD( number1 * number 2)
"""
def find_lcm_math(Number1, Number2):
    import math
    math_lcm = (Number1 * Number2) // math.gcd(Number1 , Number2)
    return f"LCM using Math package is %i"%math_lcm
```

```python
Number1 = int(input("Please enter first number "))
Number2 = int(input("Please enter second number to find LCM "))

print(find_LCM(Number1,Number2))

print(find_lcm_math(Number1,Number2))
```

# 3. Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

```python
def dec_to_bin(number):
    import math
    l = []
    myl = " "
    while number != 0:
        remain = number % 2
        remain = math.trunc(remain)
        l.append(str(remain))
        number = number // 2
    myl = myl.join(l[::-1])
    print(myl)
```

```python
def dec_to_hexa(number):
    hexa_dict = {10:'A',11:'B',12:"C",13:"D",14:'E',15:"F"}
    l = []
    myl = ""
    remainder = number
    while remainder != 0:
        remainder = number % 16
        number = number //  16
        if remainder > 9:
            l.append(hexa_dict[remainder])
        elif remainder != 0:
            l.append(str(remainder))
    myl = myl.join(l[::-1])
    print(myl)
        
```

# 4. Write a Python Program To Find ASCII value of a character?

```python
def ascii_c(character):
    return "The ASCII value of {} is {}".format(character,ord(character))
```

```python
character = str(input("Please enter only one character "))
if len(character) > 1:
    print("Only one character is allowed")
else:
    print(ascii_c(character))
```

# 2. Write a Python Program to Find HCF?

```python
def find_hcf(num1,num2):
    if num1 < num2:
        smaller = num1
    else:
        smaller = num2
    for i in range(1,smaller+1):
        if (num1%i == 0) and (num2%i == 0):
            hcf = i
    return hcf
```

```python
num1 = int(input("Please enter any number \t"))
num2 = int(input("Please enter any number \t"))
find_hcf(num1,num2)
```

# 5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

```python
def add(a,b):
    return a+b

def sub(a,b):
    return a - b

def mul(a,b):
    return a * b

def div(a,b):
    return a / b

print("Please select arthmetic operation:")
print("1.Add")
print("2.Subtract")
print("3.Multiply")
print("4.Divide")

while True:
    choice = int(input("Please select from above option(1,2,3,4) \n"))
    
    if choice in (1,2,3,4):
        input_num1 = float(input('Please enter number \t'))
        input_num2 = float(input("Please enter number \t"))
        
        if choice == 1:
            print("Addition of {} {} is {}".format(input_num1, input_num2, add(input_num1,input_num2)))
            
        elif choice == 2:
            print("Subtraction of {} {} is {}".format(input_num1, input_num2, sub(input_num1,input_num2)))
            
        elif choice == 3:
            print("Multiplication of {} {} is {}".format(input_num1, input_num2, mul(input_num1,input_num2)))
            
        elif choice == 4:
            print("Division of {} {} is {}".format(input_num1, input_num2, div(input_num1,input_num2)))
            
        break
    else:
        print("Invalid choice")
        
```
