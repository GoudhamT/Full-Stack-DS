# 1.Write a Python Program to Find the Factorial of a Number?


```python
def get_fact(n):
    m = 1
    for i in range(1,n+1):
        m = m*i
    else:
        return "The facorial  of number is",m
```

```python
n = int(input("Enter any number to find factorial \n"))
get_fact(n)
```

# 3.Write a Python Program to Print the Fibonacci sequence?

```python
def fibonacci_num(n):
    a = 1
    b = 1
    for i in range(1,n+1):
        yield a
        a,b = b,a+b
        
```

```python
n = int(input("Enter number to find Fibonacci sequence \t"))
for i in fibonacci_num(n):
    print(i)
```

# 4. Write a Python Program to Check Armstrong Number?

```python
def check_armstrng_number(n):
    a = list(str(n))
    arm_num = 0
    for i in a:
        arm_num = arm_num + int(i)**3
    if arm_num == n:
        return "Given number {} is Armstrong number".format(n)
    else:
        return "Given number {} is not an Armstrong number".format(n)
```

```python
n = int(input("Enter any number to check Armstrong "))
check_armstrng_number(n)
```

# 5 .Write a Python Program to Find Armstrong Number in an Interval?

```python
def get_armstrng_number(min_num,max_num):
    interval = list(range(min_num,max_num+1))
    armstrng_num = []
    for num in interval:
        num_list = list(str(num))
        arm_num = 0
        for i in num_list:
            arm_num = arm_num + int(i)**3
            if arm_num == num:
                armstrng_num.append(num)
    return armstrng_num
    
```

```python
get_armstrng_number(151,1540)
```

# 6. Write a Python Program to Find the Sum of Natural Numbers?

```python
def sum_natural_number(n):
    return sum(range(1,n+1))
```

```python
n = int(input("Enter any number \t"))
if n >= 1:
    print(sum_natural_number(n))
else:
    print("Number cannot be less than 1")
```

# 2. Write a Python Program to Display the multiplication Table?

```python
def display_tables(n):
    for i in range(1,11):
        print(n,' X ', i, ' = ',n*i)
```

```python
n = int(input("Enter any number you want to check multiplication table values \t"))
display_tables(n)
```
